```python
img = cv2.imread('/content/gdrive/My Drive/Images/ehna.bmp')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg, cmap='gray')
```

```python
from random import randint
```

```python
# 10% noise
noiseRatio = 10
whiteMax = noiseRatio
blackMax = noiseRatio
```

```python
noiseImg = gImg.copy()

for h in range(0, gImg.shape[0]):
  for w in range(0, gImg.shape[1]):
    val = randit(0, 100)
    if(val < whiteMax):
      noiseImg[h,w] = 255
    elif(val < blackMax):
      noiseImg[h,w] = 0

cv2_imshow(noiseImg)
```

```python
outImg2 = cv2.medianBlur(noiseImg, 5)
cv2_imshow(outImg2)
```

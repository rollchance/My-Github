```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/MyDrive/ColabIP/Images/gray.bmp')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg, cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 1)
cv2_imshow(dilateImg)
```

```python
edgeImg = dilateImg - gImg
cv2_imshow(edgeImg)
```

```python
kernel = np.ones((3,3),np.uint8)
dilateImg = cv2.dilate(gImg, kernel, iterations = 5)
cv2_imshow(dilateImg)
```

```python
img = cv2.imread('/content/gdrive/My Drive/Images(1)/abcdef.bmp')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gImg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
cv2_imshow(gImg)
#plt.imshow(gImg, cmap='gray')
```

```python
outImg = cv2.GaussianBlur(gImg, (3, 3), 1)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (3, 3), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7, 7), 1)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (7, 7), 3)
cv2_imshow(outImg)
```

```python
outImg = cv2.GaussianBlur(gImg, (15, 15), 3)
cv2_imshow(outImg)
```

```python
mean_img = cv2.blur(gImg, (5, 5))
cv2_imshow(mean_img)
```

```python
edgeImg = cv2.addWeighted(gImg, 1.0, mean_img, -1.0, 0)
cv2_imshow(edgeImg)
```

```python
out_img = cv2.addWeighted(gImg, 1.0, mean_img, 1.0, 0)
cv2_imshow(out_img)
```
